```python
import pandas as pd
```

```python
# Read the CSV files from the results folder
deepseeek = pd.read_csv('results/benchmark_results_deepseek-r1-1.5b_20250616_034747.csv')
gemma3_1b = pd.read_csv('results/benchmark_results_gemma3-1b_20250616_033649.csv') 
qwen17b = pd.read_csv('results/benchmark_results_qwen3-1.7b_20250616_035955.csv')
qwen32b = pd.read_csv('results/benchmark_results_qwen3-32b_20250616_212345.csv')
qwen30b = pd.read_csv('results/benchmark_results_qwen3-30b_20250616_224559.csv')
phi4 = pd.read_csv('results/benchmark_results_phi4-latest_20250616_233231.csv')
gemma3_27b = pd.read_csv('results/benchmark_results_gemma3-27b_20250616_233420.csv')

# Calculate the average full accuracy for each model
deepseeek_avg = deepseeek.avg_full_accuracy.mean()
gemma3_1b_avg = gemma3_1b.avg_full_accuracy.mean()
qwen17b_avg = qwen17b.avg_full_accuracy.mean()
qwen32b_avg = qwen32b.avg_full_accuracy.mean()
qwen30b_avg = qwen30b.avg_full_accuracy.mean()
phi4_avg = phi4.avg_full_accuracy.mean()
gemma3_27b_avg = gemma3_27b.avg_full_accuracy.mean()

# Print the average full accuracy for each model
print(f"DeepSeek-R1-1.5B Average Full Accuracy: {deepseeek_avg:.2f}")
print(f"Gemma3-1B Average Full Accuracy: \t{gemma3_1b_avg:.2f}")
print(f"Qwen3-1.7B Average Full Accuracy: \t{qwen17b_avg:.2f}")
print(f"Qwen3-32B Average Full Accuracy: \t{qwen32b_avg:.2f}")
print(f"Qwen3-30B Average Full Accuracy: \t{qwen30b_avg:.2f}")
print(f"Phi4 Average Full Accuracy: \t\t{phi4_avg:.2f}")
print(f"Gemma3-27B Average Full Accuracy: \t{gemma3_27b_avg:.2f}")
```
